```python
# 만약에 탄소발자국의 최소값을 구하는 거라면
#x1의 탄소발자국이 50이고 x2의 탄소발자국이 100이고 x3의 탄소발자국이 28이면


global tansoresult
global calorieresult
global proteinresult
global tansuhamoolresult
global fatresult


def objective_func_tanso(x):
    
    tansoresult=0
    for i in range(0,importantlength):
        tansoresult+=chooselist[i].tanbal*x[i]
    return tansoresult

#칼로리를 제약함수로 설정 칼로리가 2300보다 많아야한다고 설정
def inequality_constraint_calori(x):
    calorieresult=0
    for i in range(0,importantlength):
        calorieresult+=chooselist[i].calorie*x[i]
    return -calorieresult+2640

#또한 2800보다는 적어야한다고 설정
def inequality_constraint_calori2(x):
    calorieresult=0
    for i in range(0,importantlength):
        calorieresult+=chooselist[i].calorie*x[i]
    return calorieresult-2160
def inequality_constraint_protein(x):
    proteinresult=0
    for i in range(0,importantlength):
        proteinresult+=chooselist[i].protein*x[i]
    return -proteinresult+77
#단백질이 최대 110이하로 설정

def inequality_constraint_protein2(x):
    proteinresult=0
    for i in range(0,importantlength):
        proteinresult+=chooselist[i].protein*x[i]
    return proteinresult-63
    
#탄수화물이 최초 70이상으로 설정

def inequality_constraint_tansuhamool(x):
    tansuhamoolresult=0
    for i in range(0,importantlength):
        tansuhamoolresult+=chooselist[i].tansuhamool*x[i]
    return -tansuhamoolresult+550
#탄수화물이 최대 110이하로 설정
def inequality_constraint_tansuhamool2(x):
    tansuhamoolresult=0
    for i in range(0,importantlength):
        tansuhamoolresult+=chooselist[i].tansuhamool*x[i]
    return tansuhamoolresult-450


    
def inequality_constraint_fat(x):
    fatresult=0
    for i in range(0,importantlength):
        fatresult+=chooselist[i].fat*x[i]
    return -fatresult+72

def inequality_constraint_fat2(x):
    fatresult=0
    for i in range(0,importantlength):
        fatresult+=chooselist[i].fat*x[i]
    return fatresult-0




def please(chooselist):
    
    constraint1={'type':'ineq','fun':inequality_constraint_calori}
    constraint2={'type':'ineq','fun':inequality_constraint_calori2}
   
    constraint7={'type':'ineq','fun':inequality_constraint_fat}
    constraint8={'type':'ineq','fun':inequality_constraint_fat2}

    constraint11={'type':'ineq','fun':inequality_constraint_tansuhamool}
    constraint12={'type':'ineq','fun':inequality_constraint_tansuhamool2}
    constraint13={'type':'ineq','fun':inequality_constraint_protein}
    constraint14={'type':'ineq','fun':inequality_constraint_protein2}



    constraint=[constraint1,constraint2,constraint7,constraint8,constraint11,constraint12,constraint13,constraint14]
    bounds=[]
    x0=[]
    for i in range(0,importantlength):
        limit=(100,1000)
        bounds.append(limit)
        x0.append(100)
    results=minimize(objective_func_tanso,x0, method="SLSQP",bounds=bounds,constraints=constraint)
    if results.success==True:
        for i in chooselist:
            Truediet.append(i.name)
        Truetanso.append(results.fun)
        
    print(results)
    
#constraint=[constraint2,constraint4,constraint6,constraint8,constraint10,constraint12,constraint14,constraint16,constraint18,constraint20]
#음식 세개가 최소 100g은 넘어야한다고 설정 최대 200g


          
```
